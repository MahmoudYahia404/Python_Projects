```python
import pandas as pd
from os import listdir

df  = pd.read_csv(f"C:/Users/M.Yahia/Downloads/Video/dataset/BTC-USD.csv")
df
```

# Add symbol col based on file name 

```python
files = listdir("C:/Users/M.Yahia/Downloads/Video/dataset")
for i in files :
    fi = pd.read_csv(fr"C:/Users/M.Yahia/Downloads/Video/dataset/{i}", on_bad_lines='skip')
    fi["symbol"] = f"{i[:-8]}"
    fi.to_csv(fr"C:/Users/M.Yahia/Downloads/Video/dataset/{i}", index = False)
    
    
df
```

# Import all the dataset files and store in df var

```python
for i in files :
    file = pd.read_csv(f"C:/Users/M.Yahia/Downloads/Video/dataset/{i}")
    df = pd.concat([df, file])
    
df
```

# Remove duplicates rows

```python
#df[df.duplicated()]
df.drop_duplicates(inplace = True)
df
```

# Get the Crypto names file to join with the dataset

```python
cryptoNames =  pd.read_csv("C:/Users/M.Yahia/c_n.csv")
cryptoNames
```

# Add crypto name col

```python
df = pd.merge(df, cryptoNames, on='symbol', how='inner')
df
```

# Reorder the columns

```python
cols = df.columns.tolist()
cols = cols[-2:] + cols[:-2]
df = df[cols]
df
```

```python
len(df['name'].unique())
```

# filter the data by date (22jan-22sep)

```python
df = df[df['timestamp'] > '2021-12-31 00:00:00']
df['timestamp'] = pd.to_datetime(df['timestamp'])
#df['month'] = pd.DatetimeIndex(df['timestamp']).month 
#df = df[df['month'] == 4]
df
```

# Describe the dataset for highest AVG crypto volume

```python
dfDesc = df[['name', 'adjclose', 'volume']].groupby('name', sort = False).describe()
dfDesc.sort_values(by=('volume', 'mean'), ascending=False).head(10)
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

gdf = df.groupby('name', sort = False)[['adjclose', 'volume']].median().reset_index()

topVolume = gdf.sort_values(by='volume', ascending=False).head(11)
topVolume = topVolume.sort_values(by='volume')

plt.rcParams['font.size'] = '30'

fig, axis = plt.subplots(figsize=(30, 15))
g = axis.barh(topVolume['name'], topVolume['volume'] )
#axis.bar_label(g, label_type="edge")
axis.axes.xaxis.set_ticklabels([])
axis.set_title("highest 10 crypto volume")
#fig.patch.set_visible(False)
#axis.axis('off')
plt.show()
```

# View (*USDT, BTC, ETC*) price over time

```python
import plotly.offline as py
import plotly.graph_objs as go

usdt = df[df['symbol'] == 'USDT']
btc = df[df['symbol'] == 'BTC']
etc = df[df['symbol'] == 'ETC']
```

```python
py.init_notebook_mode(connected=True)
```

```python
data = [go.Candlestick(x=usdt.timestamp,
                       open=usdt.open,
                       high=usdt.high,
                       low=usdt.low,
                       close=usdt.close)]

layout = go.Layout(title='Tether to USD Chartr by date',
                   xaxis={'rangeslider':{'visible':True}})

fig = go.Figure(data=data,layout=layout)
py.iplot(fig,filename='Tether_candlestick')
```

```python
data = [go.Candlestick(x=btc.timestamp,
                       open=btc.open,
                       high=btc.high,
                       low=btc.low,
                       close=btc.close)]

layout = go.Layout(title='Bitcoin to USD Chartr by date',
                   xaxis={'rangeslider':{'visible':True}})

fig = go.Figure(data=data,layout=layout)
py.iplot(fig,filename='bitcoin_candlestick')
```

```python
data = [go.Candlestick(x=etc.timestamp,
                       open=etc.open,
                       high=etc.high,
                       low=etc.low,
                       close=etc.close)]

layout = go.Layout(title='Ethereum to USD Chartr by date',
                   xaxis={'rangeslider':{'visible':True}})

fig = go.Figure(data=data,layout=layout)
py.iplot(fig,filename='Ethereum_candlestick')
```

# the difference between close price and start price (profit/loss)

```python
plt.style.use("bmh")

fig, ax = plt.subplots(nrows = 3, ncols = 1,figsize=(20, 20 ))

ax[0].plot(usdt['timestamp'],usdt['adjclose'] - usdt['open'])
ax[1].plot(btc['timestamp'],btc['adjclose'] - btc['open'])
ax[2].plot(etc['timestamp'],etc['adjclose'] - etc['open'])

ax[0].set_title("USDT")
ax[1].set_title("BTC")
ax[2].set_title("ETC")

plt.show()
```
